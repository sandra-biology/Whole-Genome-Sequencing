```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Jan 12 22:36:40 2022
@author: trjojack

Supply annovar-annotated .vcf or .vcf.gz file as the first argument on the command line.  

ex: python3 tj_annotvcf_to_funct_patho_excel.hg38.py example.vcf.gz

Script will output three files, if all are present. (.allvars.xlsx, .functional.xlsx, .pahtogentic.xlsx)

"""

import pandas as pd
import os
import sys
import gzip

# os.path.join(os.path.dirname(__file__))

# in_file = sys.argv[1]
in_file = 'ppmi.july2018_subset92834.hg38_multianno.vcf'
print(in_file)


def open_gzvcf_file(VCF_file):
    #LOAD VCF FILE WITH NO HEADER
    main_file = gzip.open(VCF_file, 'rt', encoding='utf')
    mf_lines = main_file.readlines()
    h = 0
    for i in range(0, len(mf_lines)):
        if mf_lines[i].startswith('#CHROM'):
                   h = i
    m_chopped = mf_lines[h:]
    
    n = []
    for x in m_chopped:
        y = x.split('\t')
        n.append(y)
    
    m_final = pd.DataFrame(n)
    m_final = m_final.rename(columns=m_final.iloc[0]).drop(m_final.index[0])
    m_final.reset_index(inplace=True, drop=True)
    return(m_final)


def open_vcf_file_topchopped(VCF_file):
    #LOAD VCF FILE WITH TOP ROWS CHOPPED OFF
    main_file = open(VCF_file)
    mf_lines = main_file.readlines()
    h = 0
    for i in range(0, len(mf_lines)):
        if mf_lines[i].startswith('#CHROM'):
                   h = i
    m_chopped = mf_lines[h:]
    
    n = []
    for x in m_chopped:
        y = x.split('\t')
        n.append(y)
    
    m_final = pd.DataFrame(n)
    m_final = m_final.rename(columns=m_final.iloc[0]).drop(m_final.index[0])
    m_final.reset_index(inplace=True, drop=True)
    return(m_final)

# IMPORT VCF OR VCF.GZ FILE
if in_file.endswith('.vcf'):
    main_vcf = open_vcf_file_topchopped(in_file)
elif in_file.endswith('.vcf.gz'):
    main_vcf = open_gzvcf_file(in_file)    
else:
    print('error in running functions!!!')


def extract_INFO_col(vcf_df, list_of_info):
    #EXTRACT INFO COLUMN INTO NEW COLUMNS
    info_col_list = []
    info_titles = list_of_info
    
    for i in vcf_df.index:
        split_info = vcf_df.loc[i, 'INFO'].split(';')
        l = []
        for j in split_info:
            fullsplit=j.split('=')
            if len(fullsplit) > 1:
                pre=fullsplit[0]
                post=fullsplit[1]
                for k in info_titles:
                    if k == pre:
                        l.append(post)
        info_col_list.append(l)
    
    info_col_df = pd.DataFrame(info_col_list)
    print(f'info_col_df.columns: {list(info_col_df.columns)}')
    AA_change_list = []
    for i in info_col_df.index:
        aa_ch = info_col_df.iloc[i, 3]
        AA_change_list.append(aa_ch.split(':')[-1].replace("p.", ""))
    info_col_df.insert(4, 'X', AA_change_list)
    info_col_df.drop(3, axis=1, inplace=True)
    #print(info_col_df.columns)
    print(f'len(info_col_df.columns): {len(info_col_df.columns)}, len(info_titles): {len(info_titles)}')
    print(f'info_col_df.columns: {list(info_col_df.columns.values)}')
    info_col_df.columns = info_titles
    info_col_df['rsID from INFO'] = vcf_df['ID']
    info_col_df = info_col_df[['rsID from INFO']+list_of_info]
    return(info_col_df)
    
info_DF = extract_INFO_col(main_vcf, ['Func.refGene', 'Gene.refGene', 'ExonicFunc.refGene', \
'AAChange.refGene','CADD_phred', 'gnomAD_genome_ALL', 'CLNSIG'])


def get_pathogenic_variant_from_vcf(vcf_df, info_col_df):
    
    patho_info_indexes = []
    for i in info_col_df.index:
        x = info_col_df.loc[i, 'CLNSIG']
        if 'pathogenic' in x or 'Pathogenic' in x or 'risk factor' in x or 'Risk factor' in x:
            patho_info_indexes.append(i)
    
    patho_info_df = info_col_df.loc[patho_info_indexes, :]
    patho_info_indexes_VCF_df = vcf_df.loc[patho_info_indexes, :]
    patho_info_indexes_VCF_df = pd.concat( [ patho_info_indexes_VCF_df.iloc[:,0:7], patho_info_df, patho_info_indexes_VCF_df.iloc[:,7:] ], axis=1)
    patho_info_indexes_VCF_df.reset_index(inplace=True, drop=True)
    return(patho_info_indexes_VCF_df)

pathogenic_vcf = get_pathogenic_variant_from_vcf(main_vcf, info_DF)   


def get_functional_variant_from_vcf(vcf_df, info_col_df):
    
    functional_indexes = []
    
    for i in info_col_df.index:
        #ge = info_col_df.loc[i, 'Gene.refGene']
        fu = info_col_df.loc[i, 'Func.refGene']
        ex = info_col_df.loc[i, 'ExonicFunc.refGene']
        gnom = info_col_df.loc[i, 'gnomAD_genome_ALL']
        clnsig = info_col_df.loc[i, 'CLNSIG']
        if gnom != '.':
            gnom = float(gnom)
        else:
            gnom = 0
        #if fu in ['exonic', 'splicing'] and ex != 'synonymous_SNV' and 'benign' not in clnsig and 'Benign' not in clnsig:
        if fu in ['exonic', 'splicing'] and ex != 'synonymous_SNV' and gnom < 0.05:
            functional_indexes.append(i)
    #FILTER VCF FOR FUNTIONAL VARIANTS ONLY
    functional_vcf = vcf_df.loc[functional_indexes, :]
    #FILTER VCF INFO COLUMN FOR FUNCTIONAL VARIANTS ONLY
    funct_info = info_col_df.loc[functional_indexes, :]
    functional_var_df = pd.concat( [ functional_vcf.iloc[:,0:7], funct_info, functional_vcf.iloc[:,7:] ], axis=1)
    functional_var_df.reset_index(inplace=True, drop=True)
    return(functional_var_df)
        
functional_vcf = get_functional_variant_from_vcf(main_vcf, info_DF)


#EXPORT FUNCTIONAL AND PATHOGENIC VCF FILES
if len(pathogenic_vcf) > 0:
    pathogenic_vcf.to_excel(in_file.replace('.vcf', '.pathogenic_vars.xlsx'), index=False)
if len(functional_vcf) > 0:
    functional_vcf.to_excel(in_file.replace('.vcf', '.functional_vars.xlsx'), index=False)

#ADD INFO DF TO MAIN VCF FILE
main_vcf = pd.concat([ main_vcf.iloc[:,0:7], info_DF, main_vcf.iloc[:,7:] ], axis=1)
main_vcf.to_excel(in_file.replace('.vcf', '.allvars.xlsx'), index=False)
```

```python
# checking gnomAD_genome_ALL

#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Jan 12 22:36:40 2022
@author: trjojack

Supply annovar-annotated .vcf or .vcf.gz file as the first argument on the command line.  

ex: python3 tj_annotvcf_to_funct_patho_excel.hg38.py example.vcf.gz

Script will output three files, if all are present. (.allvars.xlsx, .functional.xlsx, .pahtogentic.xlsx)

"""

import pandas as pd
import os
import sys
import gzip

# os.path.join(os.path.dirname(__file__))

# in_file = sys.argv[1]
in_file = 'ppmi.july2018_subset92834.hg38_multianno.vcf'
print(in_file)


def open_gzvcf_file(VCF_file):
    #LOAD VCF FILE WITH NO HEADER
    main_file = gzip.open(VCF_file, 'rt', encoding='utf')
    mf_lines = main_file.readlines()
    h = 0
    for i in range(0, len(mf_lines)):
        if mf_lines[i].startswith('#CHROM'):
                   h = i
    m_chopped = mf_lines[h:]
    
    n = []
    for x in m_chopped:
        y = x.split('\t')
        n.append(y)
    
    m_final = pd.DataFrame(n)
    m_final = m_final.rename(columns=m_final.iloc[0]).drop(m_final.index[0])
    m_final.reset_index(inplace=True, drop=True)
    return(m_final)


def open_vcf_file_topchopped(VCF_file):
    #LOAD VCF FILE WITH TOP ROWS CHOPPED OFF
    main_file = open(VCF_file)
    mf_lines = main_file.readlines()
    h = 0
    for i in range(0, len(mf_lines)):
        if mf_lines[i].startswith('#CHROM'):
                   h = i
    m_chopped = mf_lines[h:]
    
    n = []
    for x in m_chopped:
        y = x.split('\t')
        n.append(y)
    
    m_final = pd.DataFrame(n)
    m_final = m_final.rename(columns=m_final.iloc[0]).drop(m_final.index[0])
    m_final.reset_index(inplace=True, drop=True)
    return(m_final)

# IMPORT VCF OR VCF.GZ FILE
if in_file.endswith('.vcf'):
    main_vcf = open_vcf_file_topchopped(in_file)
elif in_file.endswith('.vcf.gz'):
    main_vcf = open_gzvcf_file(in_file)    
else:
    print('error in running functions!!!')


def extract_INFO_col(vcf_df, list_of_info):
    #EXTRACT INFO COLUMN INTO NEW COLUMNS
    info_col_list = []
    info_titles = list_of_info
    
    for i in vcf_df.index:
        split_info = vcf_df.loc[i, 'INFO'].split(';')
        l = []
        for j in split_info:
            fullsplit=j.split('=')
            if len(fullsplit) > 1:
                pre=fullsplit[0]
                post=fullsplit[1]
                for k in info_titles:
                    if k == pre:
                        l.append(post)
        info_col_list.append(l)
    
    info_col_df = pd.DataFrame(info_col_list)
    print(f'info_col_df.columns: {list(info_col_df.columns)}')
    AA_change_list = []
    for i in info_col_df.index:
        aa_ch = info_col_df.iloc[i, 3]
        AA_change_list.append(aa_ch.split(':')[-1].replace("p.", ""))
    info_col_df.insert(4, 'X', AA_change_list)
    info_col_df.drop(3, axis=1, inplace=True)
    #print(info_col_df.columns)
    print(f'len(info_col_df.columns): {len(info_col_df.columns)}, len(info_titles): {len(info_titles)}')
    print(f'info_col_df.columns: {list(info_col_df.columns.values)}')
    info_col_df.columns = info_titles
    info_col_df['rsID from INFO'] = vcf_df['ID']
    info_col_df = info_col_df[['rsID from INFO']+list_of_info]
    return(info_col_df)
    
info_DF = extract_INFO_col(main_vcf, ['Func.refGene', 'Gene.refGene', 'ExonicFunc.refGene', \
'AAChange.refGene', 'CLNSIG'])


def get_pathogenic_variant_from_vcf(vcf_df, info_col_df):
    
    patho_info_indices = []
    for i in info_col_df.index:
        x = info_col_df.loc[i, 'CLNSIG']
        if 'pathogenic' in x or 'Pathogenic' in x or 'risk factor' in x or 'Risk factor' in x:
            patho_info_indices.append(i)
    
    patho_info_df = info_col_df.loc[patho_info_indices, :]
    patho_info_indices_VCF_df = vcf_df.loc[patho_info_indices, :]
    patho_info_indices_VCF_df = pd.concat( [ patho_info_indices_VCF_df.iloc[:,0:7], patho_info_df, patho_info_indices_VCF_df.iloc[:,7:] ], axis=1)
    patho_info_indices_VCF_df.reset_index(inplace=True, drop=True)
    return(patho_info_indices_VCF_df)

pathogenic_vcf = get_pathogenic_variant_from_vcf(main_vcf, info_DF)   


def get_functional_variant_from_vcf(vcf_df, info_col_df):
    
    functional_indices = []
    
    for i in info_col_df.index:
        #ge = info_col_df.loc[i, 'Gene.refGene']
        fu = info_col_df.loc[i, 'Func.refGene']
        ex = info_col_df.loc[i, 'ExonicFunc.refGene']
        # gnom = info_col_df.loc[i, 'gnomAD_genome_ALL']
        clnsig = info_col_df.loc[i, 'CLNSIG']
        # if gnom != '.':
        #     gnom = float(gnom)
        # else:
        #     gnom = 0
        if fu in ['exonic', 'splicing'] and ex != 'synonymous_SNV' and 'benign' not in clnsig and 'Benign' not in clnsig:
            print(f'i: {i}, fu: {fu}')
            functional_indices.append(i)
        # if fu in ['exonic', 'splicing'] and ex != 'synonymous_SNV' and gnom < 0.05:
        #     functional_indexes.append(i)
    #FILTER VCF FOR FUNTIONAL VARIANTS ONLY
    functional_vcf = vcf_df.loc[functional_indices, :]
    #FILTER VCF INFO COLUMN FOR FUNCTIONAL VARIANTS ONLY
    funct_info = info_col_df.loc[functional_indices, :]
    functional_var_df = pd.concat( [ functional_vcf.iloc[:,0:7], funct_info, functional_vcf.iloc[:,7:] ], axis=1)
    functional_var_df.reset_index(inplace=True, drop=True)
    return(functional_var_df)

print(f'info_DF columns: {info_DF.columns}')
functional_vcf = get_functional_variant_from_vcf(main_vcf, info_DF)
print(f'functional vcf:')
display(functional_vcf)


# #EXPORT FUNCTIONAL AND PATHOGENIC VCF FILES
# if len(pathogenic_vcf) > 0:
#     pathogenic_vcf.to_excel(in_file.replace('.vcf', '.pathogenic_vars.xlsx'), index=False)
# if len(functional_vcf) > 0:
#     functional_vcf.to_excel(in_file.replace('.vcf', '.functional_vars.xlsx'), index=False)

# #ADD INFO DF TO MAIN VCF FILE
# main_vcf = pd.concat([ main_vcf.iloc[:,0:7], info_DF, main_vcf.iloc[:,7:] ], axis=1)
# main_vcf.to_excel(in_file.replace('.vcf', '.allvars.xlsx'), index=False)
```
